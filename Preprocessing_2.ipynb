```python
# Need to import to use date time
from datetime import datetime, date
 
# need to import for working with pandas
import pandas as pd
 
# need to import to use pyspark
from pyspark.sql import Row
 
# need to import for session creation
from pyspark.sql import SparkSession
 
# creating the session
spark = SparkSession.builder.getOrCreate()
```

```python
df3 = spark.read.options(header='True', inferSchema='True', delimiter=',') \
  .csv("fin.csv")
```

```python
df3.printSchema()
```

```python
df4 = df3.select("match_id","batting_team","striker","runs_off_bat").groupBy("match_id","batting_team","striker").sum().orderBy("match_id","batting_team")
```
