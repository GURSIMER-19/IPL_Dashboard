```python
#Joining both the datasets
df3 = pd.merge(df, df2, on='match_id', how='left')
df3
```

```python
df_ = df3[df3['bowling_team'].isna()]
df_not = df3[df3['bowling_team'].notna()]
```

```python
#merging both datsets into final one.
df_final = pd.concat([df_not, df_])
df_final.info()
```

```python
#changing data type
df_final["date"]= pd.to_datetime(df_final["date"])
```

```python
#Sorting data by date
df_final = df_final.sort_values(by=['date'])
```
